```python
tr_gl = int(len([gl for gl in failed_crop_names if "Tr-gl" in gl]))
tr_me = int(len([me for me in failed_crop_names if "Tr-me" in me]))
tr_no = int(len([no for no in failed_crop_names if "Tr-no" in no]))
tr_pi = int(len([pi for pi in failed_crop_names if "Tr-pi" in pi]))


te_gl = int(len([gl for gl in failed_crop_names if "Te-gl" in gl]))
te_me = int(len([me for me in failed_crop_names if "Te-me" in me]))
te_no = int(len([no for no in failed_crop_names if "Te-no" in no]))
te_pi = int(len([pi for pi in failed_crop_names if "Te-pi" in pi]))

frequency_tr = [tr_gl,tr_me,tr_no,tr_pi] #Primer conjunto de datos
frequency_te = [te_gl,te_me,te_no,te_pi] #Segundo conjunto de datos

#Crear histograma para visualizar las frecuecnais

labels_1 = ['tr_gl','tr_me','tr_no','tr_pi']
labels_2 = ['te_gl','te_me','te_no','te_pi']

print(frequency_tr)
plt.figure(figsize=(10,5))

#Primer historgrama
plt.subplot(121)
plt.pie(frequency_tr, labels = labels_1, autopct='%1.1f%%')
plt.title("Frecuencia de crops fallidos en el Training")

plt.subplot(122)
plt.pie(frequency_te, labels = labels_2, autopct='%1.1f%%')
plt.title("Frecuencia de crops fallidos en el Testing")
plt.show()
```
